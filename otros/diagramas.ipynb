```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# Configuracin de fuente
plt.rcParams['font.family'] = 'Calibri'
plt.rcParams['font.size'] = 11

fig, ax = plt.subplots(figsize=(10, 12))
ax.axis('off')

# Define boxes with text and positions
boxes = [
    ("Dataset:\nMaceracin y fermentacin", (0.1, 0.9)),
    ("Preprocesamiento:\nLimpieza, normalizacin,\ntratamiento de atpicos", (0.1, 0.78)),
    ("Control estadstico de procesos:\nGrficas de control de ABV", (0.1, 0.66)),
    ("Reduccin de dimensionalidad:\nPCA de variables dinmicas", (0.1, 0.54)),
    ("Entrenamiento de modelos:\nRegresin lineal, PLS,\nRandom Forest, XGBoost", (0.1, 0.42)),
    ("Evaluacin comparativa:\nRMSE, MAE, R,\nvalidacin cruzada", (0.1, 0.30)),
    ("Interpretabilidad:\nAnlisis SHAP", (0.6, 0.54)),
    ("Despliegue de pipeline:\nPreprocesamiento y prediccin\nreproducible", (0.6, 0.30)),
]

# Draw boxes
for text, (x, y) in boxes:
    ax.add_patch(Rectangle((x, y - 0.05), 0.35, 0.09, fill=False))
    ax.text(x + 0.005, y, text, va='top', ha='left', fontsize=10)

# Vertical arrows between main sequence
for i in range(5):
    x_center = boxes[i][1][0] + 0.35 / 2
    y_start = boxes[i][1][1] - 0.05
    y_end = boxes[i + 1][1][1] + 0.045
    ax.annotate("", xy=(x_center, y_end), xytext=(x_center, y_start),
                arrowprops=dict(arrowstyle="->"))

# Arrow from PCA to Interpretabilidad
x_start = boxes[3][1][0] + 0.35
y_start = boxes[3][1][1] - 0.045
x_end = boxes[6][1][0]
y_end = boxes[6][1][1] + 0.045
ax.annotate("", xy=(x_end, y_end), xytext=(x_start, y_start),
            arrowprops=dict(arrowstyle="->"))

# Arrow from Evaluacin to Despliegue
x_start = boxes[5][1][0] + 0.35
y_start = boxes[5][1][1] - 0.045
x_end = boxes[7][1][0]
y_end = boxes[7][1][1] + 0.045
ax.annotate("", xy=(x_end, y_end), xytext=(x_start, y_start),
            arrowprops=dict(arrowstyle="->"))

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch

# Configuracin de fuente
plt.rcParams['font.family'] = 'Calibri'
plt.rcParams['font.size'] = 11

fig, ax = plt.subplots(figsize=(10, 12))
ax.axis('off')

boxes = [
    ("Dataset:\nMaceracin y fermentacin", (0.1, 0.9)),
    ("Preprocesamiento:\nLimpieza, normalizacin,\ntratamiento de atpicos", (0.1, 0.78)),
    ("Control estadstico de procesos:\nGrficas de control de ABV", (0.1, 0.66)),
    ("Reduccin de dimensionalidad:\nPCA de variables dinmicas", (0.1, 0.54)),
    ("Entrenamiento de modelos:\nRegresin lineal, PLS,\nRandom Forest, XGBoost", (0.1, 0.42)),
    ("Evaluacin comparativa:\nRMSE, MAE, R,\nvalidacin cruzada", (0.1, 0.30)),
    ("Interpretabilidad:\nAnlisis SHAP", (0.6, 0.54)),
    ("Despliegue de pipeline:\nPreprocesamiento y prediccin\nreproducible", (0.6, 0.30)),
]

# Dibujar cajas con estilo
for text, (x, y) in boxes:
    box = FancyBboxPatch(
        (x, y - 0.05), 0.35, 0.1,
        boxstyle="round,pad=0.3",
        linewidth=1.5,
        edgecolor="#21618C",
        facecolor="#EBF5FB"
    )
    ax.add_patch(box)
    ax.text(x + 0.175, y - 0.005, text, ha='center', va='center')

# Flechas principales
arrow_kwargs = dict(arrowstyle="->", color="#1B4F72", linewidth=1.5)
for i in range(5):
    x0, y0 = boxes[i][1]
    x1, y1 = boxes[i + 1][1]
    ax.annotate(
        "", 
        xy=(x1 + 0.175, y1 + 0.045), 
        xytext=(x0 + 0.175, y0 - 0.055),
        arrowprops=arrow_kwargs
    )

# Flecha de PCA a Interpretabilidad
x0, y0 = boxes[3][1]
x1, y1 = boxes[6][1]
ax.annotate(
    "",
    xy=(x1 - 0.175, y1 + 0.045),
    xytext=(x0 + 0.35, y0 - 0.055),
    arrowprops=arrow_kwargs
)

# Flecha de Evaluacin a Despliegue
x0, y0 = boxes[5][1]
x1, y1 = boxes[7][1]
ax.annotate(
    "",
    xy=(x1 - 0.175, y1 + 0.045),
    xytext=(x0 + 0.35, y0 - 0.055),
    arrowprops=arrow_kwargs
)

plt.tight_layout()
plt.show()
```
